### Задача 2.
Построить на отдельных графиках линии уровня функций $z=\ln(x^2+y)$ и $z=x^2-y^2$.

```python
# Пример 1
x, y = symbols('x y')
C = symbols('C', real=True)
def contourline(f, x, y, C):
  return sympy.Eq(f(x, y), C)
def f(x, y):
  return Log(x**2 + y)
p = plot(ylabel='y', show=False)
for item in [(-2, 'blue'), (-1, 'green'), (0, 'red'), (2, 'black'), (5, 'pink')]:
  c, col = item
  p.append(plot_implicit(contourline(f, x, y, c), line_color=col, line_width=15, show=False)[0])
p.show()
```

```python
# Пример 2
x, y = symbols('x y')
C = symbols('C', real=True)
def contourline(f, x, y, C):
  return sympy.Eq(f(x, y), C)
def f(x, y):
  return Log(x**2 - y**2)
p = plot(ylabel='y', show=False)
for item in [(-2, 'blue'), (-1, 'green'), (0, 'red'), (2, 'black'), (5, 'pink')]:
  c, col = item
  p.append(plot_implicit(contourline(f, x, y, c), line_color=col, line_width=15, show=False)[0])
p.show()
```

### Задача 6.
Проверить выполнение теоремы Эйлера для функции $f(x,y)=\frac{x}{x^2+y^2} $.
###### Указание.
Упростить выражение с помощью simplify, прежде чем сравнивать.

```python
k = symbols('k', real=True)
f = x / (x**2 + y**2)
simplify(f.subs({x: k*x, y: k*y}))
```

1/k = k^n

k^-1 = k^n

n = -1

Теперь проверим выполнение самой теоремы Эйлера:

```python
simplify(x*f.diff(x) + y*f.diff(y)) == -1 * f
```

### Индивидуальное задание
Построить линии уровня и найти полный дифференциал функции
$f(x, y)$.

$$f(x,y)=- 4 x^{2} + 3 y^{2}$$

```python
# Линии уровня
x, y = symbols('x y')
C = symbols('C', real=True)
def contourline(f, x, y, C):
  return sympy.Eq(f(x, y), C)
def f(x, y):
  return -4 * x**2 + 3 * y**2
p = plot(ylabel='y', show=False)
for item in [(-2, 'blue'), (-1, 'green'), (0, 'red'), (2, 'black'), (5, 'pink')]:
  c, col = item
  p.append(plot_implicit(contourline(f, x, y, c), line_color=col, line_width=15, show=False)[0])
p.show()
```

```python
# Полный дифференциал
z = -4 * x**2 + 3 * y**2
dx, dy = symbols('dx dy')
dz = z.diff(x)*dx + z.diff(y)*dy
dz
```
